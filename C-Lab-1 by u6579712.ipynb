```python
# Assume image is an col x row x channel ndarray.
def cvRGB2YUV2(image):
    tmp = image.transpose(0,2,1)
    convert_matrix = np.array([[0.299,0.587,0.114],[-0.168736,-0.331264,0.5],[0.5,-0.418688,-0.081312]])
    addition = np.array([0,128,128])
    tmp = convert_matrix @ tmp
    tmp = tmp.transpose(0,2,1)
    tmp = np.add(tmp,addition)
    print(tmp.shape)
    ans = tmp
    np.clip(tmp, 0, 255, out=ans)
    ans = ans.astype(np.uint8)
    return ans
```

```python
# Assume image is an col x row x channel ndarray.
def cvRGB2YUV(image):
    image_shape = image.shape
    ans = np.zeros(image_shape)
    
    for i in range(image_shape[0]):
        for j in range(image_shape[1]):
            R = image[i][j][0]
            G = image[i][j][1]
            B = image[i][j][2]
            Y = 0.299*R+0.587*G+0.114*B
            U = B-Y+128
            V = R-Y+128
            
            ans[i][j] = np.array([Y,U,V])
    np.clip(ans, 0, 255, out=ans)
    ans = ans.astype(np.uint8)
    return ans
```
