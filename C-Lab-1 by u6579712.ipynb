```python
# Assume image is an col x row x channel ndarray.
def cvRGB2YUV(image):
    tmp = image.transpose(0,2,1)
    convert_matrix = np.array([[0.299,0.587,0.114],[-0.168736,-0.331264,0.5],[0.5,-0.418688,-0.081312]])
    addition = np.array([0,128,128])
    tmp = convert_matrix @ tmp
    tmp = tmp.transpose(0,2,1)
    tmp = np.add(tmp,addition)
    print(tmp.shape)
    ans = tmp
    np.clip(tmp, 0, 255, out=ans)
    ans = ans.astype(np.uint8)
    return ans
```

```python
def nearest_centroid(centroids,point):
    distance_min = np.inf
    for i,centroid in enumerate(centroids):
        distance = float(np.linalg.norm(centroid-point))
        if(distance<distance_min):
            ans_centroid = centroid
            ans_idx = i
            distance_min=distance
    return (ans_idx+1),ans_centroid
```

```python
def clustering(image,centroid_num = 5,iteration_num = 3):
    # initialize
    np.random.seed(25)
    centroids = np.random.rand(centroid_num,3)
    centroids *= 155
    centroids += 100
    centroids = centroids.astype(np.uint8)
    
    
    ans = np.zeros(image.shape[:2])
    

    for i in range(iteration_num):
        # E
        for j,row in enumerate(image):
            for k,channels in enumerate(row):
                current_idx,current_centroid = nearest_centroid(centroids,channels)
                ans[j][k] = current_idx
    
        # M
        new_centroids = np.zeros(centroids.shape)
        assignment_log = dict()
        for m in range(centroid_num):
            assignment_log[m+1] = []
        for j,row in enumerate(image):
            for k,channels in enumerate(row):
#                 print(ans[j][k])
                assignment_log[ans[j][k]].append(channels)       
                
        new_centroids=[]
        for centroid_idx, logs in assignment_log.items():
            current_new_centroid = np.mean(logs,axis=0)
            new_centroids.append(current_new_centroid)
        centroids = np.array(new_centroids)
        
    return ans,assignment_log
    
    
```

```python
def my_gauss_filter(noisy_image,my_gaussian_kernel):
    # here we assume shape is valid and use zero padding for borders.
    #  output shape
    m = noisy_image.shape[0]-my_gaussian_kernel.shape[0]+1
    n = noisy_image.shape[1]-my_gaussian_kernel.shape[1]+1
    #  kernel shape
    (x,y) = my_gaussian_kernel.shape
    assert m>0
    assert n>0
    
    output_image = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            current_window = noisy_image[i:i+x,j:j+y]
            convolution_result = current_window * my_gaussian_kernel
            output_image[i][j] = np.sum(np.sum(convolution_result))
    return output_image
```

```python
# NOTICE: this self-design method uses inverse warping via bilinear interpolation.
def my_translation(image,x_trans,y_trans):
    # assume origin point is at bottom left corner.
    x_trans = -x_trans
    (x_shape,y_shape,_) = image.shape
    ans_image = np.zeros(image.shape,dtype='u1')
    
    # whether it's integer translation
    if int(x_trans)==x_trans and int(y_trans) == y_trans:
        print("Integer Translation Mapping...")
        for i in range(x_shape):
            for j in range(y_shape):
                if (i+x_trans) in range(x_shape) and (j+y_trans) in range(y_shape):
                    ans_image[i+x_trans][j+y_trans]=image[i][j]
    else:
        # using inverse warping & bilinear interpolation
        print("Non-integer Translation Mapping...")
        for i in range(x_shape):
            for j in range(y_shape):
                x_prev,y_prev=(i-x_trans,j-y_trans)
                if 0 < x_prev < x_shape-1 and 0 < y_prev < y_shape-1:
                    x_1 = int(np.floor(i-x_trans))
                    x_2 = int(np.ceil(i-x_trans))
                    y_1 = int(np.floor(j-y_trans))
                    y_2 = int(np.ceil(j-y_trans))
                    # neighbour pixels
                    q_11 = image[x_1][y_1]
                    q_12 = image[x_1][y_2]
                    q_21 = image[x_2][y_1]
                    q_22 = image[x_2][y_2]

                    
                    x_array = np.array([x_2-x_prev,x_prev-x_1])
                    pixel_matrix = np.array([[q_11,q_12],[q_21,q_22]])
#                     print(pixel_matrix.shape)
                    pixel_matrix = pixel_matrix.transpose(2,0,1)
                    y_array = np.array([y_2-y_prev,y_prev-y_1])

                    # there are 3 channels in the graph
                    ans_image[i][j] = (x_array @ pixel_matrix @ y_array)
#                     print(q_11,q_12,q_21,q_22)
#                     print(ans_image[i][j])
    return ans_image
```
